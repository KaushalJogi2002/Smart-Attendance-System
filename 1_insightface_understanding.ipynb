```python
import numpy as np
```

```python
import insightface
```

```python
from insightface.app import FaceAnalysis
```

```python
app_l= FaceAnalysis(name='buffalo_l',root='insightface_model',providers=['CPUExecutionProvider'])
app_l.prepare(ctx_id=0, det_size=(640, 640))
```

```python
app_sc= FaceAnalysis(name='buffalo_sc',root='insightface_model',providers=['CPUExecutionProvider'])
app_sc.prepare(ctx_id=0, det_size=(640, 640))
```

```python
import cv2
```

```python
img = cv2.imread('test_image_2.jpg')
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
results = app_l.get(img)
```

```python
print(results)
```

```python
type(results),len(results)
```

```python
results[0].keys() 
```

```python
img_copy=img.copy()
for res in results:
    x1,y1,x2,y2=res['bbox'].astype(int)
    
    
    cv2.rectangle(img_copy,(x1,y1),(x2,y2),(0,255,0),2)
    

cv2.imshow('bbox',img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
